```python
import pymongo
import datetime
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Declare the database
db = client.job_us_dice

# Declare the collection
collection = db.job_desc
```

```python

# Part I
# A dictionary that represents the document to be inserted
post = {
    'job_description': ' ',
    'job_id': ' ',
    'company': ' ',
    'date': datetime.datetime.utcnow()
}
# Insert the document into the database
# The database and collection, if they don't already exist, will be created at this point.
collection.insert_one(post)
```

```python
import os
import pandas as pd
```

```python
# Create a reference the CSV file desired
dicecsv = os.path.join("Resources", "Dice_US_jobs.csv")

# Read the CSV into a Pandas DataFrame
dice_df = pd.read_csv(dicecsv,  encoding="ISO-8859-1")

dice_df_droppedcol=dice_df.drop(columns=['advertiserurl', 'jobid', 'site_name'])
dice_df_droppedcol.head(3)
```

```python
dice_df_droppedcol.shape
```

```python
job_desc = ["uniq_id", "jobdescription", "company"]
job_desc_transformed= dice_df_droppedcol[job_desc].copy()

# Rename the column headers
job_desc_transformed = job_desc_transformed.rename(columns={"company": "company",
                                                          "jobdescription": "job_description",
                                                          "uniq_id": "job_id"})

# Clean the data by dropping duplicates and setting the index
job_desc_transformed.drop_duplicates("job_id", inplace=True)
# job_desc_transformed.set_index("job_id", inplace=True)
```

```python
import numpy as np
job_desc_transformed['job_id'].replace(' ', np.nan , inplace=True)
job_desc_transformed['job_description'].replace(' ', np.nan , inplace=True)
job_desc_transformed['company'].replace(' ', np.nan , inplace=True)


job_desc_transformed.dropna(subset=['job_id'], inplace=True)
job_desc_transformed.dropna(subset=['job_description'], inplace=True)
job_desc_transformed.dropna(subset=['company'], inplace=True)

```

```python
job_desc_transformed.shape
job_desc_transformed.head(5)
```

```python
#### Convert the dataframe into a dictionary - needs an index - so setting an index
job_desc_transformed.set_index("job_id", inplace=True)
```

```python
### Convert df to dict
job_desc_transformed.to_dict()
```
